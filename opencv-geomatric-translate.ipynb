```python
# import packages
import cv2
from matplotlib import pyplot as plt
import numpy as np
import imutils
```

```python
def plt_imshow(title, image):
	# convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
image = cv2.imread("opencv_logo.png")
plt_imshow("CV2 Image", image)
```

```python
(h, w, c) = image.shape[:3]

# display the image width, height, and number of channels to our
# terminal
print("width: {} pixels".format(w))
print("height: {}  pixels".format(h))
print("channels: {}".format(c))
```

## Translating or shifting the image
[
    [
        [1, 0, shiftX]
        [0, 1, shiftY]
    ]
]
###  x axis: negative value shift image to left, positive value shift image to right
### y axis: negative value shift image to top, positive value shift image to botton

```python
# shift the image 25 pixels to the right and 50 pixels down

M = np.float32([[1, 0, 25], [0, 1, 50]]) # matrix = m
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt_imshow("Shifted Down and Right", shifted)
```

```python
# now, let's shift the image 50 pixels to the left and 90 pixels
# up by specifying negative values for the x and y directions,
# respectively
M = np.float32([[1, 0, -50], [0, 1, -90]])
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt_imshow("Shifted Up and Left", shifted)
```

## alternate way 

```python
# use the imutils helper function to translate the image 100 pixels
# down in a single function call
shifted = imutils.translate(image, 0, 100)
plt_imshow("Shifted Down", shifted)
```
